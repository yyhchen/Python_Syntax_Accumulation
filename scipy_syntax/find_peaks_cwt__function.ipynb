### `find_peaks_cwt` 函数

`find_peaks_cwt` 是一个连续小波变换（CWT）的峰值检测器，用于找到信号中的峰值。下面是对 `find_peaks_cwt` 的详细解释：

```python
from scipy.signal import find_peaks_cwt
```

#### 参数说明：
- **`vector`**：要进行峰值检测的一维数组。
- **`widths`**：小波的宽度，可以是一个数组，表示不同尺度的小波宽度。
- **`noise_perc`**：控制峰值检测的灵敏度的参数，表示在计算局部最大值时，将高于全局最大值的百分比视为峰值。


示例中，`find_peaks_cwt` 被用于检测带有两个峰值的数据中的峰值。函数的 `widths` 参数定义了不同尺度的小波宽度，`noise_perc` 控制了峰值检测的灵敏度。

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks_cwt

# 生成带有峰值的数据
np.random.seed(0)
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x) + np.sin(2 * x) + np.random.normal(0, 0.1, size=len(x))

# 使用连续小波变换进行峰值检测
peaks = find_peaks_cwt(y, widths=np.arange(1, 10), noise_perc=0.1)

# 绘制原始数据和检测到的峰值
plt.plot(x, y, label='Data with Peaks')
plt.scatter(x[peaks], y[peaks], color='red', marker='x', label='Detected Peaks')
plt.legend()
plt.show()
````
