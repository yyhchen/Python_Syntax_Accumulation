### `savgol_filter` 函数

`savgol_filter` 是 Savitzky-Golay 滤波器的一种实现，用于平滑信号。这种滤波器的特点是在保留信号特征的同时去除噪音。下面是对 `savgol_filter` 的详细解释：

```python
from scipy.signal import savgol_filter
```

#### 参数说明：
- **`x`**：要平滑的一维数组。
- **`window_length`**：滤波器的窗口大小，表示用于拟合多项式的数据点数量。通常选择奇数。
- **`polyorder`**：用于拟合多项式的阶数。


示例中，`savgol_filter` 被用于平滑带有噪音的正弦函数。

````python
"""
    scipy包中的 savgol_filter函数, 用于平滑信号
"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# 生成一些带噪音的数据
np.random.seed(0)
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=len(x))

# 使用 Savitzky-Golay 滤波器进行平滑
# 对于window_length=15的情况
smoothed_y_15 = savgol_filter(y, window_length=15, polyorder=2)
# 对于window_length=7的情况
smoothed_y_7 = savgol_filter(y, window_length=7, polyorder=2)

# 绘制原始数据和平滑后的数据
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Noisy Data')
plt.plot(x, smoothed_y_15, label='Smoothed Data (window_length=15)')
plt.legend()
plt.title('Savitzky-Golay Filter with window_length=15')

plt.subplot(1, 2, 2)
plt.plot(x, y, label='Noisy Data')
plt.plot(x, smoothed_y_7, label='Smoothed Data (window_length=7)')
plt.legend()
plt.title('Savitzky-Golay Filter with window_length=7')
plt.tight_layout()
plt.show()
````
